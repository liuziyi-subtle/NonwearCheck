## 检查测试数据说明_汇总中的reference是否和真实的references文件夹一致

```python

manual_infos = pd.read_csv(
    '/Users/liuziyi/Documents/Lifesense/data/nonwear-check/O/raw/测试数据说明_汇总.csv')
print(manual_infos.shape)

real_reference_names = glob('/Users/liuziyi/Documents/Lifesense/data/nonwear-check/O/raw/2020*/references/*')
real_reference_names = list(map(lambda x: x.split('/')[-1], real_reference_names))
# real_reference_names_prefix = list(map(lambda x: x.split('/')[0], real_reference_names))
print(len(real_reference_names))

for index, series in manual_infos.iterrows():
    if series['ambient_light'] == 'weak':
        manual_infos.loc[index, 'ambient_light'] = 'low'
    if series['reference'] not in real_reference_names:
        position = list(map(lambda x: x.split('.')[0], real_reference_names)).index(series['reference'].split('.')[0])
        manual_infos.loc[index, 'reference'] = real_reference_names[position]

manual_infos.to_csv('/Users/liuziyi/Documents/Lifesense/data/nonwear-check/O/raw/测试数据说明_汇总_mod.csv', index=None)
```
